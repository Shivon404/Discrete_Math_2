 Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

- $E[y] =  \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot  ( \Sigma_{y_1} \,y_1 \cdot p_1) \cdot (\Sigma_{y_2} \, y_2 \cdot p_2) \cdot ( \Sigma_{y_3} \, y_3 \cdot p_3) \ldots ( \Sigma_{y_k} \, y_k \cdot p_k)$

- $E[y] = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (n \cdot p_1) \cdot (n \cdot p_2) \cdot (n \cdot p_3) \ldots  (n \cdot p_k)$

- $E[y] = n \cdot p_1 \cdot n \cdot p_2 \cdot n \cdot p_3 \ldots n \cdot p_k$

- $E[y] = n \cdot 1$
- $E[y] = n \cdot p_i$

To conclude, the expected value  E[y] equals to $\Sigma y \cdotp P(y)$, which is equal to $n \cdot p_i$ .

**<h4>Question 1:</h4>**

Given probability distribution: $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$

Expected value formula: $E[Y] = \Sigma_{y=0} ^ \infty \, y \cdot p(y)$

Step 1: Substitue the Probability Distribution into the Expected Value Formula

- $E[Y] = \Sigma_{y=0} ^ \infty \, y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$

Step 2: Factor out Constants

- $E[Y] = e^{-\lambda} \Sigma_{y=0} ^ \infty \, \frac{\lambda^y}{(y - 1)!} $

Step 3: Since $y = 0$ makes $\frac{\lambda^y}{(y - 1)!}$ undefined, we shift the index of summation to start from $y = 1$.

- $E[Y] = e^{-\lambda} \Sigma_{y=1} ^ \infty \, \frac{\lambda^y}{(y - 1)!} $

Step 4: Rewrite $\frac{\lambda^y}{(y - 1)!}$ as $\frac{\lambda^{(y-1)+ 1}}{(y - 1)!}$ 

- $E[Y] = e^{-\lambda} \Sigma_{y=1} ^ \infty \,  \frac{\lambda^{(y-1)+ 1}}{(y - 1)!}$

Step 5: Change the index of summation to start from $y = 0$ to simplify the expression.

- $E[Y] = e^{-\lambda} \Sigma_{y=0} ^ \infty \,  \frac{\lambda^{y+1}}{y!}$

Step 6: Factor out $\lambda$

- $E[Y] = e^{-\lambda} \lambda \Sigma_{y=0} ^ \infty \, \frac {\lambda^{y}}{y!}$

Step 7: Recall that $e^{\lambda} =  \Sigma_{y=0} ^ \infty \, \frac {\lambda^{y}}{y!}$

-  $E[Y] = e^{-\lambda} \lambda e^{\lambda} $

Step 8: Simplify: 

-  $E[Y] =  \lambda$

To conclude, the general formula for the expected value of the given distribution is $E[Y] =  \lambda$
